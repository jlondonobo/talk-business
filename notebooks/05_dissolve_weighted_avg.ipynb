```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
def dissolve_weighted_average(gdf: gdp.GeoDataFrame, by: str, column: str, weight: str) -> gdp.GeoDataFrame:
    return (
        gdf
        .filter([by, "geometry", column, weight])
        .assign(premultiplied=lambda df: df[column] * df[weight])
        .dissolve(by=by, aggfunc='sum')
        .reset_index()
        .assign(weighted_average=lambda df: df["premultiplied"] / df[weight])
    )
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
dissolve_weighted_average(world, "continent", "gdp_md_est", "pop_est")
```
